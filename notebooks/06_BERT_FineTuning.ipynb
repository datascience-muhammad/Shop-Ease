# Notebook 6: BERT Fine-Tuning

BERT already 'knows' how to read because it was trained on millions of books. In this notebook, we do **"Transfer Learning."**

**The Analogy**: Imagine hiring a professional translator (BERT) and giving them a quick 1-hour crash course on **ShopEase Customer Reviews**. You don't need to teach them how to read; you just need to teach them what your specific customers care about.

```python
import sys
from pathlib import Path

# Import our training engine
sys.path.append("..")
from src.train import train_bert, load_data

data_path = "../data/processed/clean_reviews.csv"
print("Training engine is fueled and ready.")
```

## 1. The Crash Course (Fine-Tuning)

We will show BERT 100 examples of our specific reviews so it can adjust its 'internal logic'. 

*Note: This is heavy work for a computer. It's like doing a thousand math problems in a few seconds!*

```python
X, y = load_data(data_path)

print("Starting the AI 'Crash Course'... ")
status = train_bert(X[:100], y[:100])
print("Training Complete! BERT is now a ShopEase expert.")
```
