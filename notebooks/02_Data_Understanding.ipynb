# Notebook 2: Data Understanding

Before we write a single line of AI code, we need to look at our data. Think of this as **"Detective Work."**

![Data Understanding](assets/data_understanding_visual.png)

### Our Mission:
1. See what a 'raw' review actually looks like.
2. Check if we have more positive reviews than negative ones (Class Imbalance).
3. Spot the 'noise' (emojis, slang, and mixed languages) that we need to clean later.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Locate and load the raw data
data_path = Path("../data/raw/raw_reviews.csv")
df = pd.read_csv(data_path)

print(f"We have {len(df)} customer reviews to analyze.")
df.head()
```

## 1. The Rating Distribution

**Plain English**: Are our customers mostly happy or mostly angry? 

If 90% of our data is '5 stars', the AI will struggle to learn what a bad review looks like. We want a healthy mix.

```python
df['rating'].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('How Many Reviews for Each Star Rating?')
plt.xlabel('Stars (1 to 5)')
plt.ylabel('Number of Customers')
plt.show()
```

## 2. Spotting the 'Noise'

Let's look at 5 random reviews. Notice the **emojis**, the **ALL CAPS**, and the **different languages** (German, French, Spanish).

The computer sees a 'happy face' emoji as a confusing symbol, not a feeling. We need to fix this in the next step.

```python
print("Random Real-World Examples:")
for review in df['review'].sample(5):
    print(f"- {review}")
```

## Summary
We've confirmed our data is messy and multilingual. This is perfect for teaching an AI how to be robust!
